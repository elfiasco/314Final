```python
import pandas as pd
import numpy as np
```

```python
full_data = pd.read_csv('/Users/kids/Desktop/2022 Pitches.csv')
full_data.head()
```

```python
reduced_data = full_data[['pitcher', 'batter', 'release_speed', 'release_pos_x', 'release_pos_z', 'pfx_x', 'pfx_z', 'release_spin_rate', 'release_extension', 'spin_axis', 'description']]
reduced_data["description"] = np.where(reduced_data["description"] == "swinging_strike", 1, 0)
reduced_data = reduced_data.rename(columns = {'description':'swinging_strike'})
```

```python
reduced_data['numPitches']=reduced_data.groupby(['pitcher'])['pitcher'].transform('count')
```

```python
clean_data = reduced_data[reduced_data['numPitches']>=100]
clean_data = clean_data.drop(columns=['numPitches'])
clean_data.head()
```

```python
shuffled_data = clean_data.sample(frac=1)
splitPt = round(len(shuffled_data)*0.8)
pitches = shuffled_data.iloc[:splitPt, :]
hidden = shuffled_data.iloc[splitPt:, :]
pitches.head()
```

```python
pitches.to_csv('/Users/kids/Desktop/pitches.csv')
hidden.to_csv('/Users/kids/Desktop/hidden.csv')
```
